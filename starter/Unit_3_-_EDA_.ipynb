#### Fill in blanks with a brief description of each plot type:

> visualizing the Probability Density at different values in a continuous variable.

> Bar charts are one of the many techniques used to present data in a visual form so that the reader may readily recognize patterns or trends. Bar charts usually present categorical variables, discrete variables or continuous variables grouped in class intervals.

> a line plot  s a graph that uses lines to connect individual data points
- Univariate scatter plot
> plots different observations/values of the same variable corresponding to the index/observation number
- Violin Plot
> A violin plot depicts distributions of numeric data for one or more groups using density curves.
- Box Plot
> A box plot is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum. The term "box plot" comes from the fact that the graph looks like a rectangle with lines extending from the top and bottom.
- Pie Chart
> A pie chart is the most common way used to visualize the numerical proportion occupied by each of the categories.
- Rug Plot
> A Rug Plot would be another way to display a distribution as an adjunct to your primary plot.  A Rug Plot is simply a projection of raw data points onto a particular axis, without involving the binning process of a histogram.

### Plot for 'age' column

```python
# create plot for age column
```

```python
# sets figure size
plt.figure(figsize=(10,4))

# establishes figure with data and plot type
ax = sns.histplot(data=df, x='age', palette="viridis", bins=8)

# sets x and y labels
plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)

# sets title
plt.title("Manager Ages in Survey", fontsize=15);
```

Equal amount of managers in each field

##### Insights:

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
heat=pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level

plt.figure(figsize=(16,10))

ax = sns.heatmap(heat, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
```

##### Insights:

The darker the colors, the more salary people are making because of their higher education levels in every field.
